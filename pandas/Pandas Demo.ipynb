## Corey's Approach

```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
python_df = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'Country':'NumOfRespndnts', 'LanguageWorkedWith':'KnowsPython'}, inplace=True)
python_df
```

```python
python_df['%KnowsPython'] = (python_df['KnowsPython']/python_df['NumOfRespndnts']) * 100
python_df
```

```python
python_df.sort_values(by='%KnowsPython', ascending=False, inplace=True)
```
